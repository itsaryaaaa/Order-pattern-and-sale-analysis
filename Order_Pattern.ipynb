```python
df.columns = df.columns.str.lower().str.replace(' ', '_')
```

```python
df.head()
```

```python
df['discount']=df['list_price']*df['discount_percent']*0.01
df.head()
```

```python
df['sale_price']=df['list_price']+df['discount']
df['profit']=df['sale_price']-df['cost_price']
df.head()
```

```python
df.describe()
df.columns.tolist()
#order_date is object 
```

```python
#dropping unnecessary columns
df['order_date']=pd.to_datetime(df['order_date'],format="%Y-%m-%d")
df.info()
#df.dtypes
```

```python
#connecting jupyter notebook to PostgreSQL usinh username, passwords, host, port, database
from sqlalchemy import create_engine
username= 'postgres' 
password= 'Password'
host= 'localhost'
port= '5432'
database= 'order_pattern_and_analysis'
engine=create_engine(f"postgresql+psycopg2://{username}:{password}@{host}:{port}/{database}")

#loading data into postgresql
table_name="order_pattern"
df.to_sql(table_name, engine, if_exists="replace", index=False) #no index column is stored in sql table
print(f"Data successfully loaded into table'{table_name}' in database'{database}',")
```
